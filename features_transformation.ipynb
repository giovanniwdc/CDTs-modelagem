## **1. Morphing a Set of Points**

```python
def show_morphing(Dx, Dy, x, y, xlabel = 0, ylabel = 0):
    
    # creating the grid (x,y)
    n, m = len(x), len(y)
    
    xl = np.linspace(np.min(x), np.max(x), n)
    yl = np.linspace(np.min(y), np.max(y), m)
    
    X, Y = np.meshgrid(xl, yl)
    
    # computing the vector components 
    fx = []
    fy = []
    
    for xi in xl:
        for yi in yl:
            
            fx.append(Dx(xi, yi))
            fy.append(Dy(xi, yi))

    # plot settings
    xl = xlabel if xlabel else Dx.__name__
    yl = ylabel if ylabel else Dy.__name__        
    
    fig = plt.figure(figsize = (8,8))

    plt.quiver(X, Y, fx, fy, color='r')
    
    plt.xlabel(xl)
    plt.ylabel(yl)
```

```python
x = np.linspace(0, 1, 50)
y = np.linspace(0, 1, 50)
```

```python
Dx = lambda x, y: x*y
Dy = lambda x, y: x
```

```python
show_morphing(Dx, Dy, x, y)
```

```python
def circle(x0, y0, r, n, seed = 42):
    
    np.random.seed(42)
    
    x = np.zeros(n)
    y = np.zeros(n)
    
    count = 0
    
    while count < n - 1:
        
        xl = np.random.uniform(x0-r, x0+r, 1)
        yl = np.random.uniform(y0-r, y0+r, 1)
        
        if (xl-x0)**2+(yl-y0)**2 <= r**2:
            
            x[count] = xl
            y[count] = yl
            
            count += 1
    return x, y
```

```python
def Dx(t, v):
    return -t**2*np.sign(t)

def Dy(t, v):
    return -v**2*np.sign(v)
```

```python
x, y = circle(0, 0, 1, 9)
```

```python
plt.plot(x, y, 'o')
```

```python
show_morphing(Dx, Dy, x, y)
```

```python
plt.plot([Dx(xi, yi) for xi, yi in zip(x,y)], [Dy(xi, yi) for xi, yi in zip(x,y)], 'o')
```
