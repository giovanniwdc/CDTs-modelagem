# Small Worlds Networks

## 1. Characterization

### A small world networks, is a non directed graph that has a high number of loops and a low distance degree for any two nodes. It can be constructed by using, initially, a regular lattice with nodes of $2k$ degrees, where $k$ is the number of neighbors in each direction, creating a symmetric network. In sequence, we permute every pair of edge with a probability $p$.

### One good example of a small world network are the social network, where the probability of three friends be friend of each other, pair to pair, is high (high number of loops) and the distance between two people is low (low distance degree). 

## 2. Creation

```python
np.random.seed(1)
random.seed(1)
p = 0.2

for x0 in range(n -1):
    for y0 in range(x0+1, n):
        if adj_m[x0][y0] == 1:
            if random.random() < p:
                
                while True:
                    new_node = np.random.randint(n)
                    
                    if adj_m[x0][new_node] != 1 and x0 != new_node:
                        
                        adj_m[x0][y0] = 0 
                        adj_m[y0][x0] = 0

                        adj_m[x0][new_node] = 1 
                        adj_m[new_node][x0] = 1

                        break
```
